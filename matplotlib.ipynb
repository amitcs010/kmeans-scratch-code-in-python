```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
x=np.arange(1,50,0.1)
y=np.sin(x)
plt.plot(x,y,':',color='blue',label='sin')
plt.plot(x,np.cos(x),'--',color='yellow',label='cos')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('graph')
plt.legend(loc='best')
plt.show()
```

```python
p=np.random.chisquare(1,100)
r=np.arange(-2,10,0.1)
plt.hist(p,r)
plt.show()
```

```python
a=['India','Australia','Newzeland','England','Pakistan']
s=[13,14,11,12,11]
plt.pie(s,labels=a,explode=(0.1,0,0,0,0.5),autopct='%.2f',shadow=True)
plt.show()
```

```python
a=['India','Australia','Newzeland','England','Pakistan']
a1=[0,1,2,3,4]
s=[13,14,11,12,11]
p=plt.subplot(2,2,1)
p.plot(a1,s)
p1=plt.subplot(2,2,2)
p1.scatter(a1,s)
p2=plt.subplot(2,2,3)
p2.bar(a,s)
p3=plt.subplot(2,2,4)
p3.pie(s,labels=a)
plt.show()
```

```python
p=plt.subplot(111)
p.plot(a1,s)
p.annotate("ideal point",(1,14),xytext=(0.2,0.3),textcoords="axes fraction",arrowprops=dict(color='blue'))
```
